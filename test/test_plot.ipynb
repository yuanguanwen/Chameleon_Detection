```python
import numpy as np
import matplotlib.pyplot as plt

s2eV = (6.582119569e-16)#	// Hz to eV

# data = np.loadtxt("output_T_vs_E.dat")
data = np.loadtxt("T_spectrum_ring_5to10.dat")
data2 = np.loadtxt("T_spectrum_ring_1to100.dat")
E = data[:, 0]
Tval = data[:, 1]*1e3/s2eV   # convert [eV/eV] to [s-1 keV-1]
E2 = data2[:, 0]
Tval2 = data2[:, 1]*1e3/s2eV   # convert [eV/eV] to [s-1 keV-1]

Bspec = np.loadtxt("B_spectrum_1e2.dat")
Bx, By = Bspec[:, 0], Bspec[:, 1]*1e3/s2eV
Bspec2 = np.loadtxt("B_spectrum_1e2_low.dat")
Bx2, By2 = Bspec2[:, 0], Bspec2[:, 1]*1e3/s2eV
Tspec = np.loadtxt("T_spectrum_1e2.dat")   # very fast
Tx, Ty = Tspec[:, 0], Tspec[:, 1]*1e3/s2eV

plt.plot(Bx/1e3, By, label="B_{up}", linestyle='-', color='orange')
plt.plot(Bx2/1e3, By2, label="B_{low}", linestyle='--', color='orange')
plt.plot(Tx/1e3, Ty, label="T", linestyle='-')
plt.plot(E2/1e3, Tval2, label="T_{ring=0.01}", linestyle='--')
plt.plot(E/1e3, Tval, label=" T_{ring=0.5}", linestyle=':')
plt.xlabel("Energy $w$ [eV]")
plt.ylabel(r"$T_{\mathrm{solarIntg\_x}}$ [Bg$^{-2}$]")
plt.xlim(1e-2, 20)
plt.ylim(1e14, 1e22)
plt.xscale("log")  
plt.yscale("log")  
plt.grid(True)
plt.legend()
plt.title("Chameleon Solar Production vs Energy")
plt.tight_layout()
plt.show()
```
